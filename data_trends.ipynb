# About

This notebook deals with visualizing the data of the titanic dataset in an attempt to determine which features may be relevant for prediciting survival. We heavily make use of data visualization so that we can spot significant trends. 

# Prep and Clean Data

```python
## Setup
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
%matplotlib inline


data_file = '/home/andy/Projects/Kaggle/Titanic/Data/train.csv'
df = pd.read_csv(data_file,header=0)

###### Clean Data #########################
# 1) Sex | females:0 , males:1
df['Gender'] = df['Sex']
df['Sex'] = df['Sex'].map( {'female': 0, 'male': 1} ).astype(int)

# 2) Family (total number of family members)
df['Family'] = df['SibSp'] + df['Parch'] #Sibsp = no. of siblings/spouses, Parch = no. spouses/children

# 3) Embarked - S:0, Q:1, C:2
df.loc[ (df.Embarked.isnull()), "Embarked"] =  df.Embarked.mode()[0] #fill null values with mode
df['Embarked'] = df['Embarked'].map( {'S':0, 'Q':1, 'C':2} ).astype(int) #convert to integer values

# 4) Age 
#Determine median ages based on sex and pclass
median_ages = np.zeros((2,3))

for i in range(0, 2):
    for j in range(0, 3):
        median_ages[i,j] = df[ (df['Sex'] == i) & (df['Pclass'] == j+1) ]\
        ['Age'].dropna().median()

#Fill in null values of Age
for i in range(0, 2):
    for j in range(0, 3):
        df.loc[ ( df.Age.isnull() ) & ( df['Sex'] == i ) & (df['Pclass'] == j+1), \
        "Age"] = median_ages[i,j]

# 5) Fare Price 	

#Fill in Fare Brackets 
#0: (0-10) | 1: (10-20) | ... | 4: (30-40) | 5:(40+)
for i in range(6):
    df.loc[ df.Fare >= 10*i, 'Fare_Bracket'] = i

###############################
# Drop Irrelevant Features
###############################
df = df.drop(['PassengerId','Name', 'Ticket', 'Cabin', 'SibSp','Parch','Fare'], axis=1)
```

# Uncovering Trends - using Data Visualization

$\textbf{Survival based on Gender}$

```python
"""
Gender
"""

fig = plt.figure(figsize=(20,8), dpi=1600 )
a=0.8

axes_1 = fig.add_subplot(2,1,1) 
female = df.Survived[df.Sex == 0].value_counts()
female.plot(kind='bar', label='female',color='red', alpha=a)
axes_1.set_xlim(-1, len(female))
plt.legend(loc='best')

axes_2 = fig.add_subplot(2,1,2) 
male= df.Survived[df.Sex == 1].value_counts()
male.plot(kind='bar', label='male', color='green', alpha=a)
axes_2.set_xlim(-1, len(male))
plt.legend(loc='best')

plt.show()
```

$\textbf{Surivival based on Gender and Class}$:

```python
""" 
Consider Class and Gender
"""

#group functionality
df.groupby(['Pclass','Gender']).Survived.sum().plot(label = 'survived', kind='barh')
plt.legend(loc='best')

#compare survival rates based on pclass and gender
death_counts = pd.crosstab([df.Pclass, df.Gender], df.Survived)
death_counts.plot(kind='bar', stacked=False, color=['black','gold'], grid=False)

#compare survived:death ratios (another way of visualization)
death_counts_ratio = death_counts.div(death_counts.sum(1).astype(float), axis=0)
death_counts_ratio.plot(kind='barh',stacked=True,color=['black','green'])

```

$\textbf{Survival: Age and Gender}$

```python
"""
Consider Age and Gender
"""

#classify based on age (10s,20s,30s,...)
import math
df['Age_bracket']=df['Age'].map( lambda x: math.floor(x/10)*10) 

#Grab info for age_bracket and survival based on gender
df1 = df[df['Gender'] == 'female'][['Age_bracket', 'Survived']] #females
df2 = df[df['Gender']=='male'][['Age_bracket','Survived']] #males

survival_female = pd.crosstab(df1.Age_bracket, df1.Survived)
survival_male = pd.crosstab(df2.Age_bracket, df2.Survived)

survival_female_ratio = survival_female.div(survival_female.sum(1).astype(float), axis=0).plot(kind='barh',stacked=True,color=['red','blue'])
survival_male_ratio = survival_male.div(survival_female.sum(1).astype(float), axis=0).plot(kind='barh',stacked=True,color=['black','green'])
plt.legend(loc='best')
```

$\textbf{Survival: Location and Family Size}$

```python

"""
Consider location
"""
location_counts = pd.crosstab([df.Embarked],df.Survived)
location_counts.plot(kind = 'bar', stacked = False, color = ['orange', 'blue', 'purple'])
location_counts_ratio = location_counts.div(location_counts.sum(1).astype(float),axis=0)
location_counts_ratio.plot(kind = 'barh', stacked = True, color = ['orange', 'blue', 'purple'] )

"""
Consider family size
"""

family = pd.crosstab([df.Family],[df.Survived])
family_ratio = family.div(family.sum(1).astype(float), axis=0)
family_ratio.plot(kind='barh',stacked=True)
```

$\textbf{Concluding Insights}$

$\textbf{Gender:}$
Clearly, a bigger proportion of females survived than males.

$\textbf{Wealth / Fareclass:}$
We see that wealthier individuals (1st and 2nd class) have a much better chance of survival than 3rd class. One possible explanation is that that first and second class patrons received access to lifeboats first. 

$\textbf{Age:}$
There seems to be no glaring trends for age so we will ignore it for now. 

$\textbf{Location:}$
There does not seem to be any major correlations between the location and survival. For sake of simplicity, we do not deem this feature as relevant for our initial model. In the future, we may be interested in quantitatively determining this correlation. 

$\textbf{Family Size:}$
Oddly enough, we see that if you have between 0-3 family members, your survival rate is substantially more likely than if you have 4+ family members. This is a bit unintuitive, but it seems that it is relevant enough of a trend to include in our prediction.
